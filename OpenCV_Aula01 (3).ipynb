```python
from google.colab.patches import cv2_imshow
import cv2

cap = cv2.VideoCapture('video.mp4')
while cap.isOpened():
    ret, image = cap.read()

    if not ret:
      break

    cv2_imshow(image) # Note cv2_imshow em vez de cv2.imshow

    cv2.waitKey(1) & 0xff

cv2.destroyAllWindows()
cap.release()
```
